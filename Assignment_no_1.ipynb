```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("images.jpg",cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

```python
negative=255-img
cv2_imshow(negative)
```

```python
c = 100 / np.log(1 + np.max(img))
log_transformed = c * np.log(1 + img)
log_transformed = np.array(log_transformed, dtype=np.uint8)

cv2_imshow(log_transformed)
print("Log Transformation")
```

```python
gamma = 1.9  # Try values like 0.4, 0.7, 1.5 etc
gamma_corrected = np.array(255 * (img / 255) ** gamma, dtype=np.uint8)

cv2_imshow(gamma_corrected)
print(f"Power-Law (Gamma) Transformation (={gamma})")
```

```python
# Apply simple thresholding
threshold_value = 180
_, thresholded = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)

cv2_imshow(thresholded)
print(f"Thresholding at value = {threshold_value}")
```

```python
# Clip pixel values between 50 and 200
min_clip = 50
max_clip = 200

clipped = np.clip(img, min_clip, max_clip)
clipped = np.array(clipped, dtype=np.uint8)

cv2_imshow(clipped)
print(f"Clipping Transformation (Values clipped between {min_clip} and {max_clip})")
```
